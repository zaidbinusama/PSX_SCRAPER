AFTER CHANGING RUN ALL CELLS

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup


arr =[]
# Initialize WebDriver
driver = webdriver.Chrome()

# Open the web page


# 
# 
# THIS IS THE URL OF THE PAGE YOU WANT TO SCRAPE. CHANGE IT TO THE PSX URL OF THE PAGE YOU WANT TO SCRAPE
company = 'PIAHCLA'
driver.get('https://dps.psx.com.pk/company/'+company)

# Click on the "5Y" button
five_year_button = driver.find_element(By.CSS_SELECTOR, 'div.chart__range__item[data-name="5Y"]')
five_year_button.click()

# Wait until the overlay becomes visible
element_to_hover_over = WebDriverWait(driver, 15).until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "rect.overlay"))
)



for i in range(-240, 220):
    # Perform hover action at specific X, Y coordinates within the overlay
    try:
        hover = ActionChains(driver).move_to_element_with_offset(element_to_hover_over, i, 20)
        hover.perform()

        # Wait for the pop-up to appear
        WebDriverWait(driver, 5).until(EC.visibility_of_element_located((By.CLASS_NAME, 'popover')))

        # Scrape the pop-up content
        pop_up_html = driver.page_source
        soup = BeautifulSoup(pop_up_html, 'html.parser')

        # Extract the required data from the pop-up
        pop_up_div = soup.find_all('div', {'class': 'popover'})
        arr = arr + pop_up_div
        print(i)

    except:
        print("missed "+str(i))


# hover = ActionChains(driver).move_to_element_with_offset(element_to_hover_over, -240, 20)
# hover.perform()


# pop_up_html = driver.page_source
# soup = BeautifulSoup(pop_up_html, 'html.parser')


# pop_up_div = soup.find_all('div', {'class': 'popover'})
# print(pop_up_div)

```

CHECKING ARRAY ELEMENTS

```python
print(arr[457].find('div', {'class': 'popover__date'}).text)
print(arr[457].find_all('div', {'class': 'popover__value'})[0].text)
print(arr[457].find_all('div', {'class': 'popover__value'})[1].text)
```

CREATING DICTIONARY

```python
hashtable = {}
for i in range(0, len(arr)):
    hashtable[arr[i].find('div', {'class': 'popover__date'}).text] = [arr[i].find_all('div', {'class': 'popover__value'})[0].text, arr[i].find_all('div', {'class': 'popover__value'})[1].text]
    
```

DUMPING DICTIONARY INTO FILE

```python
import json
driver.quit()
filename=company+'.json'
with open(filename, 'w') as fp:
    json.dump(hashtable, fp, indent=4)

print("done")
```
